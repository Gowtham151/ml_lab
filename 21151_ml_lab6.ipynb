```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
data = pd.read_csv('Malayalam_Char_Gabor.csv')
X = data.iloc[:, 2].values.reshape(-2, 1) 
y = data.iloc[:, 3].values.reshape(-3, 1)  
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print('Mean Squared Error:', mse)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
data = pd.read_csv('Malayalam_Char_Gabor.csv')

X_features = data.iloc[:, 2:].values
y_target = data.iloc[:, 1].values.reshape(-1, 1)
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_features, y_target)
model_coefficients = linear_regression_model.coef_
model_intercept = linear_regression_model.intercept_
y_predicted = linear_regression_model.predict(X_features)
mean_squared_err = mean_squared_error(y_target, y_predicted)
r_squared = r2_score(y_target, y_predicted)

print('Model Coefficients:', model_coefficients)
print('Model Intercept:', model_intercept)
print('Mean Squared Error:', mean_squared_err)
print('R-squared:', r_squared)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_csv('Malayalam_Char_Gabor.csv')
X = data.iloc[:, 4].values.reshape(-1, 1)  
y = data.iloc[:, 6].values.reshape(-1, 1)  

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('LocalEnergy')
plt.ylabel('MeanAmplitude')
plt.title('Linear Regression Model')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('Malayalam_Char_Gabor.csv')
X = data.iloc[:, 2:].values  
y = data.iloc[:, 1].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=53)

tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)
mse_tree = mean_squared_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)
knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)
print('Regression Tree - Mean Squared Error:', mse_tree)
print('Regression Tree - R-squared:', r2_tree)
print('k-NN Regressor - Mean Squared Error:', mse_knn)
print('k-NN Regressor - R-squared:', r2_knn)
```
